```python
def lpr_plot(scan,point,datpath,lprpath):
    datfile=os.path.join(datpath,f'asdc_data_{scan:03}_{point}_lpr.csv')
    dflpr = pd.read_csv(datfile)
    t = analysis.LPRData(dflpr)
    #t.clip_current_to_range()
    t.plot(fit=True)
    ax=[]
    ax =plt.gca()
    plt.ticklabel_format(style='sci', axis='x', scilimits=(0,0))
    plt.ylabel('potential (V vs. Ag/AgCl)')
    plt.title(f'LPR:{scan:03}:{point}--pH={t.pH_med}--LPR={np.round(t.polarization_resistance)} (ohm)--R^2={np.round(t.r_value,decimals=4)}')
    plt.savefig(os.path.join(lprpath,f'lpr_fitted__{scan:03}_{point}.tiff'))
    plt.close()
    return t
```

```python
def tafel_plot(scan,subscan,datpath,tafelpath):
    datfile=os.path.join(datpath,f'asdc_data_{scan:03}_{subscan}_tafel.csv')
    dftafel = pd.read_csv(datfile)
    t = analysis.TafelData(dftafel)
    t.clip_current_to_range()
    t.plot(fit=True,lsv_threshold=.8,tafel_binsize=.01 )
    ax=plt.gca()
    loggap=np.abs(np.log10(t.tafel_data["cathodic"]["j0"])-np.log10(t.tafel_data["anodic"]["j0"]))
    plt.title(f'Scan {scan}:{subscan}--I_corr={t.i_corr:.3e}--I_range={loggap:.2}')#--redchi^2={t.model.redchi:.3e}
    filename=os.path.join(tafelpath,f'tafel_fitted__{scan:03}_{subscan}.tiff')
    #if os.path.isfile(filename):
    #    os.remove(filename)
    plt.savefig(filename)
    plt.show()
    plt.close()
    return t
```
